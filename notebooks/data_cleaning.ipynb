```python
#since this is just exploration and cleaning we only need the essentials
import pandas as pd
import numpy as np
import json
```

## Call in the Dataset üìÉ

```python
# Using pd.read_json() didn't work.
# We will make our own data frame.
data = []
with open('../raw_data/unpacked_json_files/yelp_academic_dataset_business.json', 'r') as file:
    for line in file:
        try:
            json_data = json.loads(line)
            data.append(json_data)
        except json.JSONDecodeError:
            # Handle any invalid JSON lines here
            pass

# Create a DataFrame from the loaded JSON objects
business_df = pd.DataFrame(data)

# Display the DataFrame
business_df
```

```python
# trying to see if there are any tokyo places.
tokyo_df = business_df[business_df['city']=='Tokyo']
```
